# **Pregunta - 2:**

Realiza el proceso de limpieza. Aplica el preprocesamiento que consideres adecuado.











```

# ******* Inlcuye a continuación todas las líneas de código y celdas que requieras: ***********
# Separamos la información:
 #     La "X" serán los datos de entrada, los comentarios.
 #     La "Y" será la variable de salida, la evaluación.
 # Ambos, X y Y son "Series"

X = df.review     # Serie de strings
Y = df.label      # Serie de enteros 0s y 1s

def clean_tok(doc):

    # Convertir a minúsculas : Unifica el formato del texto.
    doc = doc.lower()

    # Manejo de contracciones comunes
    doc = re.sub(r"(\w+)'(t|ve|m|ll|s|re|d)\b", r"\1\2", doc)

    # Tokenización : Divide el texto en palabras
    tokens = re.split(r'\W+', doc)

    # Eliminar stopwords : Remueve palabras irrelevantes
    negwords = [ 'no', 'nor', 'not', 'ain', 'aren', "aren't", 'don', "don't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]
    mystopwords = [word for word in stopwords.words('english') if word not in negwords]
    tokens = [word for word in tokens if word not in mystopwords]

    # Eliminar caracteres que no sean alfabéticos (mantiene solo letras) : Se limpia el texto
    tokens = [re.sub(r'[^a-z]', '', word) for word in tokens]

    # Filtrar tokens con longitud mayor a 1 : Evita incluir caracteres individuales o palabras sin significado.
    tokens = [word for word in tokens if len(word) > 1]

    return tokens

# Aplicamos el proceso de limpieza y tokenización:

Xclean = [clean_tok(x) for x in X]

# *********** Aquí termina la sección de agregar código *************
```

# **Pregunta - 6:**


#### **Diccionario clave-valor de palabras del diccionario y vectores embebidos.**

```
import pickle  # Importar módulo para guardar archivos binarios

def obtener_embedding(palabra):
    """
    Codifica una palabra en su representación embebida.
    """
    embedding = model.encode(palabra)
    return np.array(embedding) if embedding is not None else None  # Convertir a NumPy array

def generar_diccionario_embeddings(vocabulario):
    """
    Genera un diccionario con las palabras y sus vectores embebidos.
    """
    diccionario_embeddings = {}

    for palabra in vocabulario:
        embedding = obtener_embedding(palabra)
        if embedding is not None:  # Evitar valores None
            diccionario_embeddings[palabra] = embedding
    return diccionario_embeddings

# Usar las palabras filtradas del diccionario midicc
diccionario_embeddings = generar_diccionario_embeddings(midicc.keys())

# Guardar el diccionario en un archivo pickle
nombre_archivo = "diccionario_embeddings.pkl"
with open(nombre_archivo, "wb") as archivo:
    pickle.dump(diccionario_embeddings, archivo)

print(f"Diccionario de embeddings guardado en {nombre_archivo}")

# Mostrar ejemplos
print("Ejemplo de embeddings generados:")
for palabra, embedding in list(diccionario_embeddings.items())[:5]:
    print(f"{palabra}: {embedding[:5]}...")  # Muestra los primeros valores del embedding
```

# **Pregunta - 7:**



Generamos los vectores embebidos a partir de los conjuntos de entrenamiento, validación y prueba.

Los llamaremos trainEmb, valEmb y testEmb, respectivamente.

```
import numpy as np

def obtener_vector_promedio(comentario, diccionario_embeddings, dimension_modelo):
    """
    Calcula el vector promedio de un comentario basado en los embeddings de sus palabras.
    Si el comentario no tiene palabras en el diccionario, devuelve un vector de ceros.
    """
    embeddings = [diccionario_embeddings[w] for w in comentario if w in diccionario_embeddings]
    if embeddings:
        return np.mean(embeddings, axis=0)  # Promediar los vectores embebidos
    else:
        return np.zeros(dimension_modelo)  # Vector de ceros si no hay palabras conocidas

# Dimensión del modelo de HuggingFace seleccionado
dimension_modelo = model.get_sentence_embedding_dimension()

# Generar los conjuntos de entrenamiento, validación y prueba
trainEmb = np.array([obtener_vector_promedio(comentario, diccionario_embeddings, dimension_modelo) for comentario in train_x])
valEmb = np.array([obtener_vector_promedio(comentario, diccionario_embeddings, dimension_modelo) for comentario in val_x])
testEmb = np.array([obtener_vector_promedio(comentario, diccionario_embeddings, dimension_modelo) for comentario in test_x])

```
