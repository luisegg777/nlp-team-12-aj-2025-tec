```python
# Buscando usar GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Usando dispositivo: {device}")
```

```python
import json
# Escribir resultado en json
def escribir_json(datos, nombre_archivo):
  with open(nombre_archivo, "w", encoding="utf-8") as f:
    json.dump(datos, f, ensure_ascii=False, indent=2)

# Leer resultado de json
def leer_json(nombre_archivo):
  with open(nombre_archivo, "r", encoding="utf-8") as f:
    return json.load(f)
```

### Selección del modelo de Whisper a usar #####
En este caso en particular he decidido comenzar con el modelo whisper-large-v3 porque es el modelo mas completo que hay. Si bien se tiene el modelo turbo, esa velocidad de entrega podría comprometer en algo los resultados.

Por ello decidí tomar el mas lento pero que me asegure la mayor precisión en los resultados.

Posiblemente luego corra todo con el modelo turbo para ver la diferencia en los resultados.

```python
# Llamar al modelo de whisper large-v3 para mejor precisión
model_id = "openai/whisper-large-v3"    # mejor modelo

# Cargamos el modelo:
processor = AutoProcessor.from_pretrained(model_id)
modelo = AutoModelForSpeechSeq2Seq.from_pretrained(model_id).to(device)
```

```python
# Creamos el pipeline para el reconocimiento de voz
pipe = pipeline("automatic-speech-recognition",
    model=modelo,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device=device
    )
```

```python
# Extraemos el texto de todos los audios:
transcripciones = {}

# Procesar cada audio
archivos_mp3 = sorted([f for f in os.listdir(".") if f.endswith(".mp3")])
for archivo in archivos_mp3:
    ruta_audio = archivo
    print(f"Procesando {archivo}...")

    audio, sr = librosa.load(ruta_audio)
    result = pipe(audio,
      return_timestamps=True,
      generate_kwargs={"language": "es"}
    )

    transcripciones[archivo] = result["text"]

for nombre, texto in transcripciones.items():
    print(f"\n {nombre}:\n{texto}\n")

# Guardar las transcripciones
escribir_json(transcripciones, "transcripciones.json")
```

# **Ejercicio 2b:**

* #### **Eliminar el inicio y final comunes de los textos extraídos de cada fábula.**

* #### **Sugerencia:** Pueden guardar esta información en un archivo tipo JSON, para que al estar probando diferentes opciones en los ejercicios siguientes, puedan recuperar rápidamente la información de cada video/fábula.

```python
# Funcion para limpiar la parte inicial y final de la fabula

import re

def limpiar_fabula(texto):
    patron = r"F[aá]bula n[uú]mero\s+\d+[^\w\n]*(.*?)(?=Fin de (la )?f[aá]bula|$)"

    match = re.search(patron, texto, re.IGNORECASE | re.DOTALL)

    if match:
        return match.group(1).strip()
    else:
        return ""
```

```python
transcripciones = leer_json("transcripciones.json")
fabulas_limpias = {nombre: limpiar_fabula(texto) for nombre, texto in transcripciones.items()}

for nombre, fabula in fabulas_limpias.items():
    print(f"\n{nombre}:\n{fabula}\n")

escribir_json(fabulas_limpias, "fabulas_limpias.json")
```

#Justificación de limpieza

Por lo visto en el contenido de las fábulas, son textos cortos y cada palabra carga con el significado del contexto en el que se encuentra.

> En una iteración previa no realice un borrado de stopwords y me percaté que al realizar la obtención de las palabras clave por medio de LDA, muchos stop words estaban tomando reelevancia, lo que me dió a entender que no estaba bien el proceso de limpieza. Por ello, en esta siguiente iteración decidí ser un poco mas agresivo y eliminar stopwords en español.

> En una segunda iteración, verifique que incluso con esa limpieza los resultados no eran tan cercanos a lo esperado, investigando encontré otra libreria llamada spacy que en teoría realiza varias labores en conjunto como lemantizar y tokenizar y es buena con textos en español.
Con ello, se creó una segunda función *limpiar_texto_spacy*.

En ese sentido, considero debemos realizar una limpieza no agresiva, pero si considerada para evitar un mal calculo de palabras clave.

Por ello decidí hacer lo siguiente:
- Convertir todo el texto a minusculas
- Quitar todo tipo de salto de linea, tabulación que no deba estar (es poco probable que exista, pero para tener un estandar lo consideré)
- Borrar todos los caracteres que no sean alfanuméricos
- Eliminar espacios multiples
- Tokenizar el texto obtenido
- Eliminar stop words en español

La segunda función realiza labores similares:
-Lemmantizar
-Quitar stop words
-Quitar puntuaciones
-Quitar espacios de mas
-Quitar palabras menores o iguales a 2 de longitud

Terminé usando la segunda

```python
# Funciones de limpieza de texto
import re
import nltk
from nltk.corpus import stopwords
nltk.download('punkt_tab')
nltk.download('stopwords')
!python -m spacy download es_core_news_sm
import spacy
nlp = spacy.load("es_core_news_sm")

# Funcion responsable de limpiar las fabulas transcritas y tokenizarla al final de la limpieza
def limpiar_texto(texto):
  stopwords_es = set(stopwords.words('spanish'))
  texto = texto.lower()
  texto = re.sub(r'[\n\t\r]+', ' ', texto)
  texto = re.sub(r"[^a-zA-ZáéíóúÁÉÍÓÚñÑ0-9\s]", "", texto)
  texto = re.sub(r'\s{2,}', ' ', texto)
  tokens = nltk.word_tokenize(texto.strip(), language='spanish')
  tokens = [t.lower() for t in tokens if t.lower() not in stopwords_es]
  return tokens

def limpiar_texto_spacy(texto):
    doc = nlp(texto.lower())
    tokens = [
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct and not token.is_space and len(token.text) > 2
    ]
    return tokens
```

```python
# Limpiar y tokenizar las fabulas
fabulas_limpias = leer_json("fabulas_limpias.json")
fabulas_limpias_tokenizadas = {nombre: limpiar_texto_spacy(fabula) for nombre, fabula in fabulas_limpias.items()}

fabulas_limpias_tokenizadas.items()
```

```python
# Incluyan a continuación todas las celdas (de código o texto) que deseen...
import gensim
import gensim.corpora as corpora

def CrearDicionarioYCorpus(fabulas_limpias_tokenizadas):
  # Crear diccionario
  diccionario = corpora.Dictionary(fabulas_limpias_tokenizadas.values())

  # Crear corpus
  corpus = [diccionario.doc2bow(tokens) for tokens in fabulas_limpias_tokenizadas.values()]

  return diccionario, corpus
```

```python
# Crear modelo LDA
from gensim import models

diccionario, corpus = CrearDicionarioYCorpus(fabulas_limpias_tokenizadas)

lda_model = models.LdaModel(
    corpus=corpus,
    id2word=diccionario,
    num_topics=len(corpus),
    random_state=42,
    passes=10,
    per_word_topics=True
)

# Extraer palabras clave por fábula
palabras_clave_por_fabula = {}
for idx, topic in lda_model.show_topics(num_topics=len(corpus), num_words=20, formatted=False):
    palabras = [palabra for palabra, prob in topic]
    fabula_id = list(fabulas_limpias_tokenizadas.keys())[idx]
    palabras_clave_por_fabula[fabula_id] = palabras

# Mostrar resultado
for fabula, palabras in palabras_clave_por_fabula.items():
    print(f"\n{fabula}:\n - " + ", ".join(palabras))
```

# **Ejercicio 5a y 5b:**

* #### **5a: Mediante el LLM que hayan seleccionado, generar un único enunciado que describa o resuma cada fábula.**

* #### **5b: Mediante el LLM que hayan seleccionado, generar tres posibles enunciados diferentes relacionados con la historia de la fábula.**

* #### **Sugerencia:** En realidad los dos incisos a y b se pueden obtener con un solo prompt que solicite la información y el formato correspondiente para cada una de estas partes. Por ejemplo, para cada fábula la salida puede ser un primer enunciado genérico que resume o describe dicha temática; seguido de tres enunciados, cada uno hablando sobre una situación o parte diferente de la fábula.

```python
# Generar llamado a OpenAI para obtener el enunciado y los tres posibles enunciados
from google.colab import userdata
from openai import OpenAI
import openai

openaimodel="gpt-4-turbo"

# Metodo de realizar la conexión a openAI y retornar el cliente a utilizar
def ConectarOpenAI():
  api_key = userdata.get("openapikey")

  if not api_key:
    raise ValueError("API key no encontrada en los secretos")

  client = OpenAI(api_key = api_key)

  return client

def LlamaServicioOpenAI(client, prompt):
  try:
    response = client.chat.completions.create(
        model=openaimodel,
        messages=[
            {"role": "system", "content": "Eres un experto en análisis literario y resumen de textos."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )
    texto_generado = response.choices[0].message.content.strip()
    return texto_generado

  except Exception as e:
    print(f"Error con {nombre_fabula}: {e}")
    return "ERROR"

# Este método se encarga de tomar las palabras claves y generar un enunciado principal
# y otros tres posibles enunciados relacionados a las palabras claves
def ObtenerEnunciadosFabulasDesdeOpenAI(palabras_clave):
  client = ConectarOpenAI()
  prompt = (
      f"Tengo esta lista de palabras clave sobre una fábula:\n"
      f"{', '.join(palabras_clave)}\n\n"
      "Vas a realizar la siguiente accion:\n"
      "1. A partir de estas palabras, genera UN enunciado breve que resuma el mensaje de la fábula.\n"
      "2. Luego, genera TRES posibles subtemas distintos relacionados con el contenido de la fábula.\n"
      "El formato será el siguiente:\n"
      "Enunciado principal:\n"
      "Enunciado 1:\n"
      "Enunciado 2:\n"
      "Enunciado 3:\n"
      "No agreges mas enunciados ni quites. No agreges texto adicional a la respuesta.\n"
  )

  return LlamaServicioOpenAI(client, prompt)


# Este método se encarga de tomar los enunciados generados y compararlo
# con el tema de la fabula original
# retorna un procentaje de exactitud
def CompararEnunciadosConFabula(fabula, palabras_clave):
  client = ConectarOpenAI()
  print(f"\nFabula:\n{fabula}\n\nPalabras clave:\n{palabras_clave}")
  prompt = (
      f"Esta es la fábula:\n"
      f"{fabula}\n\n"
      "Y estos son los enunciados obtenidos en base a una lista de palabras claves:\n"
      f"{', '.join(palabras_clave)}\n\n"
      "Vas a realizar la siguiente accion:\n"
      "Vas a evaluar cada enunciado en base al contenido de la fabula y vas a calcular un porcentaje de precisión.\n"
      "Al final agregaras unos comentarios de cuales fueron las razones de la diferencia entre el enunciado y la fábula"
      "El formato será el siguiente:\n"
      "EP: XX%\n"
      "E1: XX%\n"
      "E2: XX%\n"
      "E3: XX%\n"
      "Balance General: XX%"
      "Comentarios:"
      "No agreges mas información de la descrita en el formato.\n"
  )

  return LlamaServicioOpenAI(client, prompt)
```

```python
# Llamar funciones para generar los enunciados
enunciados_por_fabula = {}
for nombre_fabula, palabras in palabras_clave_por_fabula.items():
  enunciados_por_fabula[nombre_fabula] = ObtenerEnunciadosFabulasDesdeOpenAI(palabras)
  print(f"{nombre_fabula}:\n{enunciados_por_fabula[nombre_fabula]}\n")
```

```python
# Vamos a grabar los enunciados en un json
escribir_json(enunciados_por_fabula, "enunciados_por_fabula.json")
```

```python
# Como paso adicional, buscamos comparar la similitud semantica de los enunciados
# y de la fabula para calcular que tan eficiente fue el proceso de limpieza
from sentence_transformers import SentenceTransformer, util
import re
import pandas as pd

modelo_similitud = SentenceTransformer('hiiamsid/sentence_similarity_spanish_es')

fabulas_limpias = leer_json("fabulas_limpias.json")
enunciados_por_fabula = leer_json("enunciados_por_fabula.json")

modelo_similitud = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

def extraer_enunciados(texto):
    """Extrae los enunciados del texto concatenado."""
    patron = r"(Enunciado principal|Enunciado 1|Enunciado 2|Enunciado 3):\s*(.*?)\n(?=(Enunciado|\Z))"
    matches = re.findall(patron, texto, flags=re.DOTALL)
    return {etiqueta: contenido.strip() for etiqueta, contenido, _ in matches}

def calcular_similitud(fabula, enunciado_str):
    """Calcula la similitud entre la fábula y cada uno de los enunciados extraídos."""
    enunciados = extraer_enunciados(enunciado_str)
    fabula_embedding = modelo_similitud.encode(fabula, convert_to_tensor=True)
    resultados = {}
    for tipo, texto in enunciados.items():
        enunciado_embedding = modelo_similitud.encode(texto, convert_to_tensor=True)
        similitud = util.cos_sim(fabula_embedding, enunciado_embedding).item()
        resultados[tipo] = round(similitud * 100, 2)  # porcentaje
    return resultados

evaluaciones = {}
for nombre, fabula in fabulas_limpias.items():
    if nombre in enunciados_por_fabula:
        enunciado_texto = enunciados_por_fabula[nombre]
        evaluaciones[nombre] = calcular_similitud(fabula, enunciado_texto)

df_resultados = pd.DataFrame.from_dict(evaluaciones, orient="index")
df_resultados["Balance General"] = df_resultados.mean(axis=1).round(2)
print(df_resultados)

```

```python
# Hacemos una segunda comparación, pero esta vez usando OpenAI
# para saber que tan preciso fue el resultado en base a las palabras claves
enunciados_por_fabula = leer_json("enunciados_por_fabula.json")
fabulas_limpias = leer_json("fabulas_limpias.json")

# Llamar a OpenAI para evaluar la exactitud de los resultados
for nombre_fabula, fabula in fabulas_limpias.items():
  resultado = CompararEnunciadosConFabula(fabula, enunciados_por_fabula[nombre_fabula])
  print(f"{nombre_fabula}:\n{resultado}\n")

```
